```
timeframe = "2011-10-27 2012-01-22"
dates = df.pop("date")

campaign_start = list(dates).index("2011-11-24")
campaign_end = campaign_start + 5 * 7

brands = ['Apple', 'Blackberry', 'HTC', 'Samsung']
brands_synonyms = ['Apple,iPhone', 'Blackberry', 'HTC,Evo', 'Samsung,Galaxy']
```

```
sim_matrix_dict = {}
for i, date in enumerate(dates):
    tmp_mat = create_similarity_matrix(
        df=df,
        i=i,
        brands=brands,
        brands_synonyms=brands_synonyms)
    sim_matrix_dict[date] = tmp_mat
```

```
app_sim = pd.DataFrame(index=dates)
for brand in brands:
    app_sim[brand] = get_similarity("Apple", brand, sim_matrix_dict)
```

```
def create_dataset(app_sim, dates,after_var, campaign_start, campaign_end):
    ad_period_df = pd.DataFrame()

    """For each iteration in this loop, we are defining a new dataframe
       (brand_df) for each brand. After defining the new dataframe, we append
       it to the main dataframe (ad_period_df)"""
    
    # Loop through each Brand
    for brand in ["Samsung", "HTC", "Blackberry"]:   
        brand_df = pd.DataFrame({
            "app_sim": list(app_sim[brand]),         # Distance from brand to Apple stored in column "app_dist"
            "brand_pair": [brand] * len(dates),        # Brand {Samsung, HTC, Blackberry} stored in "brand_pair" 
            "day": dates,                              # All 88 dates stored in "day"
            "during": [0] * len(dates),                # During defined as 1 ONLY if brand is Samsung and date is during the marketing campaign (0 otherwise)
            "after": [0] * len(dates)                  # After defined as either week_1, week_2, week_3 or week_4 ONLY if brand is Samsung and date is after the marketing campaign (0 otherwise)
        })
  
        if brand == "Samsung":                         # Conditional Statement to help define the during and after variables
            brand_df.loc[campaign_start:campaign_end, "during"] = 1
            brand_df.loc[campaign_end+1:len(dates), "after"] = list(after_var)
        ad_period_df = pd.concat([ad_period_df, brand_df], axis=0, ignore_index=True)   # Join "brand_df" at the end of "ad_period_df"

    return ad_period_df
```

```
ad_df= create_dataset(app_sim, dates, after_var, campaign_start, campaign_end)
```
